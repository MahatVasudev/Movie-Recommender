```python
usernames = []
counter = 0
duplicate_counter = 0
# while counter is greater than 0 
while counter < len(auth_user) or duplicate_counter <= 10000:
    # word1 randomizes names
    choice1 = choices(names + verbs,k=6)
    choice2 = choices(verbs + nouns_games, k=6)
    choice3 = choices(nouns_games + names,k=6)
    word1 = choice(choice1)
    # word2 randomizes verbs
    word2 = choice(choice2)
    # word1 randomizes nouns/games
    word3 = choice(choice3)
    # index for each word
    words4 = choices(puntuations,k=3)
    index = list(range(6))
    shuffle(index) #which is then shuffle
    word_soup = [word1, word2, word3]
    word_soup.extend(words4) #stored in a wordsoup
    username = f'{word_soup[index[0]]}{word_soup[index[1]]}{word_soup[index[2]]}'#finnally username is created
    if username == "___":
        continue
    if username not in usernames:
        usernames.append(username)
        counter = counter +  1
        print(f"counter = {counter:,}, duplicate = {duplicate_counter:,}",end='\r')
    else:
        duplicate_counter += 1
```

```python
from surprise import Dataset, Reader, SVD
from sklearn.model_selection import cross_validate  
reader = Reader()

data = Dataset.load_from_df(ratings[['userId','movieId','rating']],reader)
svd = SVD()
train_set = data.build_full_trainset()
svd.fit(train_set)
def collaborative(uid,sampling_num=1000, seed=None, n=10):
    sample_data = movies[['id','new title']].sample(n=sampling_num, random_state=seed)
    listin = []
    for id in sample_data['id']:
        sii = svd.predict(uid=uid,iid=id).est
        listin.append(sii)

    sek = pd.Series(listin, index= sample_data['new title'],name='predicted_rating')
    sek = sek.reset_index().sort_values('predicted_rating',ascending=False).head(n)
    return sek
```
