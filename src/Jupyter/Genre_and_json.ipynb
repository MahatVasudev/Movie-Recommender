```python
def genre_maker():
    listin = []
    for i in range(len(df)):
        listin.append(pd.json_normalize(df.loc[i,'genres']))

    genres = pd.concat(listin)
```

```python
df['genres'] = df['genres'].str.replace("\'","\"")
```

```python
genres = pd.json_normalize(df['genres'].apply(json.loads).apply(lambda x:x), )
```

```python
genres.isna().sum()
```

```python
new_merge = genres.merge(df['id'], on=df.index, ).drop('key_0',axis=1)
```

```python
new_merge
```

**Since there are 8 numbered columns, it probably means there are maximum of 8 different types of genres in a movie** 

To check..

```python
for i in range(len(new_merge.columns)-1):
    new_merge.rename({i: f'genre_{i+1}'},inplace=True,axis=1)
```

```python
new_merge = new_merge.reset_index(drop=True)
```

```python
new_merge.info()
```

According to the above information only three movies have 8 different genres

```python
setti = set()
for i in range(len(new_merge.columns) -1):
    for j in range(len(new_merge)):
        vec = new_merge.loc[j][f'genre_{i+1}']
        if vec == None:
            continue
        else:
            setti.add(vec['name'])
```

```python
setti
```

```python
print(f"There are in total {len(setti)} types of genres")
```

```python
dummy = pd.DataFrame(new_merge['id'],columns=['id'] + list(setti))
```

```python
dummy = dummy.fillna(0)
```

```python
def fill_genres():
    for i in range(len(dummy)):
        for j in range(len(new_merge.columns)-1):
            vec = new_merge.loc[i][f'genre_{j+1}']
            if vec == None:
                break
            else:
                dummy.loc[i,vec['name']] = 1
```

```python
fill_genres()
```

```python
dummy.to_csv('./data/genre_utility_matrix_with_movies.csv',index = False)
```

```python
user_dummy = pd.merge(ratings, new_merge, left_on='movieId', right_on='id', how= 'left')
```

```python
user_dummy = user_dummy.fillna('N/R')
```

```python
utility_matrix_users = pd.DataFrame(columns=list(setti), index=user_dummy['userId'].unique())
```

```python
utility_matrix_users
```

```python
utility_matrix_users = utility_matrix_users.fillna(0)
```

```python
utility_matrix_users=utility_matrix_users.astype(float)
```

```python
from itertools import product
```

```python
def fill_user_utitlity():
    for i in range(1,len(utility_matrix_users)+1):
        print(f"Completed user id's: {i:,}", flush=True, end='\r')
        vec = user_dummy[user_dummy['userId'] == i].reset_index(drop=True)
        for j,k in product(range(8),range(len(vec))):
            dicti = vec.loc[k][f'genre_{j+1}']
            if dicti == 'N/R':
                continue
            else:
                utility_matrix_users.loc[i,dicti['name']] = utility_matrix_users.loc[i][dicti['name']]+vec.loc[k]['rating']
```

```python
fill_user_utitlity()
```
