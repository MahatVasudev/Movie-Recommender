```python
df = pd.read_csv("movies_cleaned_v5.csv")
```

```python
still_left = df[(df['poster_done'] == 'Still Blurred') | (df['poster_done'] == 'Blurred') |(df['poster_done'] == None)]
```

```python
still_left = df[(df['poster_link'].isna()) & (df['poster_done'] != True)]

driver = webdriver.Chrome()
driver.implicitly_wait(10)
driver.get("https://www.google.com/")
for j,i in enumerate(still_left.index,1):
    try: 
        title = df.at[i,"new title"]
        search_bar = driver.find_element(By.XPATH,'//div/textarea[contains(@class,"gLFyf")]')
        search_bar.send_keys("What is the movie poster for {}".format(title))
        search_bar.submit()
        images_button = driver.find_element(By.XPATH,'//div[text()="Images"]').click()
        # filter_button = driver.find_element(By.XPATH,'//span[text()="Filter"]')
        # filter_button.click()

        if j == 1:
            time.sleep(5)
        try:
            image = driver.find_element(By.XPATH,'//div[contains(@class,"F0uyec")]/div/div[contains(@class,"H8Rx8c")]')
            image.click()
        except:
            image = driver.find_element(By.XPATH,'//div[contains(@class,"fR600b")]')
            image.click()
        try:
            image_big = driver.find_element(By.XPATH,'//div[contains(@class,"p7sI2")]/a/img[contains(@class,"sFlh5c")]')
            df.at[i,"poster_link"] = image_big.get_attribute("src")
            df.at[i,"poster_done"] = True
        except:
            df.at[i,"poster_done"] = "still not done"
        cancel_button = driver.find_element(By.XPATH,'//div[contains(@class,"M2vV3")]')
        cancel_button.click()
    except:
        df.at[i,"poster_done"] = "error"
        break
```
