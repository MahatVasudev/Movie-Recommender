```python
stemmer = nltk.SnowballStemmer(language='english')
```

```python
def clean(text):
    text = str(text).lower()
    text = ' '.join(i for i in text.split() if i not in STOPWORDS)
    text = ' '.join(stemmer.stem(i) for i in text.split())

    return text


df2['overview_cleaned'] = df2['overview'].apply(clean) 
```

```python
df2 = df2.reset_index()
```

```python
indices = pd.Series(df2.index, index= df2['new title'])
indices = indices[~indices.index.duplicated(keep='first')]
```

```python
nlp = spacy.load("en_core_web_lg")
```

```python
def similarity(string: str, sample_size = 10000, n = 10, seed=2432):
    overview1 = df2.iloc[indices.loc[string]]['overview_cleaned']
    doc1 = nlp(overview1)
    # w1 = set(ss for word in overview1 for ss in wordnet.synsets(word))
    samples = indices.sample(sample_size, random_state=seed)
    listin = []
    for i in samples.index:
        overview12 = df2.iloc[indices.loc[i]]['overview_cleaned']
        doc2 = nlp(overview12)
        # w2 = set(ss for word in overview12 for ss in wordnet.synsets(word))
        sim = doc1.similarity(doc2)
        # sim = (wordnet.wup_similarity(s1,s2) for s1,s2, in product(w1,w2))
        listin.append(sim)

    series = pd.Series(listin, index= samples.index, name="similarity")
    series = series.reset_index().sort_values("similarity",ascending=False).head(n)
    return series
```

```python
similarity("Inception (2010)",sample_size=5000, seed=221)
```

```python
from surprise import Dataset, Reader, SVD
```

```python
from sklearn.model_selection import cross_validate  
```

```python
reader = Reader()

data = Dataset.load_from_df(ratings[['userId','id','rating']],reader)
```

```python
svd = SVD()

```

```python
train_set = data.build_full_trainset()
```

```python
svd.fit(train_set)
```

```python
def collaborative(uid,sampling_num=1000, seed=None, n=10):
    sample_data = df2[['id','new title']].sample(n=sampling_num, random_state=seed)
    listin = []
    for id in sample_data['id']:
        sii = svd.predict(uid=uid,iid=id).est
        listin.append(sii)

    sek = pd.Series(listin, index= sample_data['new title'],name='predicted_rating')
    sek = sek.reset_index().sort_values('predicted_rating',ascending=False).head(n)
    return sek
```

```python
collaborative(3,sampling_num=5000,seed=1445)
```
